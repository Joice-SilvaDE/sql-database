```python
import matplotlib.pyplot as plt
import pandas as pd
import plotly as pl
import seaborn as sns
```

```python
european_happiness = pd.read_csv("european_happiness.csv")
MEA_happiness = pd.read_csv("african_happiness.csv")
```

```python
european_happiness['Region'] = "European"
MEA_happiness['Region'] = "Middle East and Africa"
```

***EUROPE***

```python
# Display the top 10 countries European 
top_10_european = european_happiness.sort_values(by='Happiness Score', ascending=False).head(10)

# Reindex the DataFrame
top_10_european.reset_index(drop=True, inplace=True)
top_10_european
```

```python
# pivot top eu
pivot_top_eu = top_10_european.reset_index().pivot(index='Country', columns='Region', values='Happiness Score')
sorted_pivot_top_eu = pivot_top_eu.sort_values(by=pivot_top_eu.columns[0], ascending=False)
```

```python
sorted_pivot_top_eu = sorted_pivot_top_eu.sort_values(by= 'European', ascending = False)
sorted_pivot_top_eu
```

```python
#top_10_european

# Plotting the sorted pivot table as a bar plot
sorted_pivot_top_eu.plot(kind='bar', figsize=(10, 6), alpha=0.5)
plt.title('Top 10 Europe - Happiness Scores by Country and Region')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45)
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Display the lower 10 countries European 
lower_10_european = european_happiness.sort_values(by='Happiness Score', ascending=False).tail(10)

# Reindex the DataFrame
lower_10_european.reset_index(drop=True, inplace=True)
lower_10_european
```

```python
# pivot table for lower european_happiness
pivot_lower_eu = lower_10_european.reset_index().pivot(index='Country', columns='Region', values='Happiness Score')
sorted_pivot_lower_eu = pivot_lower_eu.sort_values(by=pivot_lower_eu.columns[0], ascending=False)
sorted_pivot_lower_eu = sorted_pivot_lower_eu.sort_values(by= 'European', ascending = False)
sorted_pivot_lower_eu
```

```python
sorted_pivot_lower_eu = sorted_pivot_lower_eu.sort_values(by= 'European', ascending = False)
```

```python
# graph for lower european_happiness
```

```python
# Plotting the sorted pivot table as a bar plot
sorted_pivot_lower_eu.plot(kind='bar', figsize=(10, 6), alpha=0.5)
plt.title('Lower 10 Europe - Happiness Scores by Country and Region')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45)
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```

***MEA***

```python
# Display the top 10 countries MEA 
top_10_mea = MEA_happiness.sort_values(by='Happiness Score', ascending=False).head(10)

# Reindex the DataFrame
top_10_mea.reset_index(drop=True, inplace=True)
```

```python
# pivot top mea
pivot_top_mea = top_10_mea.reset_index().pivot(index='Country', columns='Region', values='Happiness Score')
sorted_pivot_top_mea = pivot_top_mea.sort_values(by=pivot_top_mea.columns[0], ascending=False)
sorted_pivot_top_mea
```

```python
# graft top mea

# Plotting the sorted pivot table as a bar plot
sorted_pivot_top_mea.plot(kind='bar', figsize=(10, 6), alpha=0.5)
plt.title('Top 10 MEA - Happiness Scores by Country and Region')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45)
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Display the lower 10 countries MEA 
lower_10_mea = MEA_happiness.sort_values(by='Happiness Score', ascending=False).tail(10)

# Reindex the DataFrame
lower_10_mea.reset_index(drop=True, inplace=True)
```

```python
# pivot_lower_mea
pivot_lower_mea = lower_10_mea.pivot_table(index='Country', columns='Region', values='Happiness Score', aggfunc='mean')

sorted_pivot_lower_mea = pivot_lower_mea.sort_values(by=pivot_lower_mea.columns[0], ascending=False)
sorted_pivot_lower_mea
```

```python
sorted_pivot_lower_mea = sorted_pivot_lower_mea.sort_values(by= 'Middle East and Africa', ascending = False)
```

```python
# graft lower mea

# Plotting the sorted pivot table as a bar plot
sorted_pivot_lower_mea.plot(kind='bar', figsize=(10, 6), alpha=0.5)
plt.title('Lower 10 MEA - Happiness Scores by Country and Region')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45)
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```
