```python
import matplotlib.pyplot as plt
import pandas as pd
import plotly as pl
import seaborn as sns
```

```python
european_happiness = pd.read_csv("european_happiness.csv")
MEA_happiness = pd.read_csv("african_happiness.csv")
```

```python
european_happiness
```

```python
european_happiness['Region'] = "European"
MEA_happiness['Region'] = "Middle East and Africa"
```

***EUROPE***

```python
# Display the top 10 countries European 
top_10_european = european_happiness.sort_values(by='Happiness Score', ascending=False).head(10)

# Reindex the DataFrame
top_10_european.reset_index(drop=True, inplace=True)
top_10_european
```

```python
# pivot top eu
pivot_top_eu = top_10_european.reset_index().pivot(index='Country', columns='Region', values='Happiness Score')
sorted_pivot_top_eu = pivot_top_eu.sort_values(by=pivot_top_eu.columns[0], ascending=False)
```

```python
sorted_pivot_top_eu = sorted_pivot_top_eu.sort_values(by= 'European', ascending = False)
sorted_pivot_top_eu
```

```python
# pivot table for bottom european_happiness
pivot_bottom_eu = bottom_10_european.reset_index().pivot(index='Country', columns='Region', values='Happiness Score')
sorted_pivot_bottom_eu = pivot_bottom_eu.sort_values(by=pivot_bottom_eu.columns[0], ascending=False)
sorted_pivot_bottom_eu = sorted_pivot_bottom_eu.sort_values(by= 'European', ascending = False)
sorted_pivot_bottom_eu
```

```python
sorted_pivot_bottom_eu = sorted_pivot_bottom_eu.sort_values(by= 'European', ascending = False)
```

***MEA***

```python
# Display the top 10 countries MEA 
top_10_mea = MEA_happiness.sort_values(by='Happiness Score', ascending=False).head(10)

# Reindex the DataFrame
top_10_mea.reset_index(drop=True, inplace=True)
```

```python
# pivot top mea
pivot_top_mea = top_10_mea.reset_index().pivot(index='Country', columns='Region', values='Happiness Score')
sorted_pivot_top_mea = pivot_top_mea.sort_values(by=pivot_top_mea.columns[0], ascending=False)
sorted_pivot_top_mea
```

```python
# Display the bottom 10 countries MEA 
bottom_10_mea = MEA_happiness.sort_values(by='Happiness Score', ascending=False).tail(10)

# Reindex the DataFrame
bottom_10_mea.reset_index(drop=True, inplace=True)
```

```python
# pivot_bottom_mea
pivot_bottom_mea = bottom_10_mea.pivot_table(index='Country', columns='Region', values='Happiness Score', aggfunc='mean')

sorted_pivot_bottom_mea = pivot_bottom_mea.sort_values(by=pivot_bottom_mea.columns[0], ascending=False)
sorted_pivot_bottom_mea
```

```python
sorted_pivot_bottom_mea = sorted_pivot_bottom_mea.sort_values(by= 'Middle East and Africa', ascending = False)
```

```python
# Calculate the average happiness index
average_top_10_eu = sorted_pivot_top_eu['European'].mean()
average_top_10_eu
```

```python
# Calculate the average happiness index
average_bottom_10_eu = sorted_pivot_bottom_eu['European'].mean()
average_bottom_10_eu
```

```python
# Calculate the average happiness index
average_top_10_mea = sorted_pivot_top_mea['Middle East and Africa'].mean()
average_top_10_mea
```

```python
# Calculate the average happiness index
average_bottom_10_mea = sorted_pivot_bottom_mea['Middle East and Africa'].mean()
average_bottom_10_mea
```

```python
# new graph top comparison
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Set theme
sns.set_theme(style="whitegrid")

# Create separate DataFrames for top and bottom comparisons
top_comparison = pd.concat([
    sorted_pivot_top_eu.assign(Category='Top EU'),
    sorted_pivot_top_mea.assign(Category='Top MEA')
]).reset_index()

bottom_comparison = pd.concat([
    sorted_pivot_bottom_eu.assign(Category='bottom EU'),
    sorted_pivot_bottom_mea.assign(Category='bottom MEA')
]).reset_index()

# Melt DataFrames for better visualization
top_melted = top_comparison.melt(id_vars=['Country', 'Category'], var_name='Region', value_name='Happiness Score')
bottom_melted = bottom_comparison.melt(id_vars=['Country', 'Category'], var_name='Region', value_name='Happiness Score')

# Create figure for top comparison
fig, ax = plt.subplots(figsize=(12, 8))
sns.barplot(data=top_melted, x='Country', y='Happiness Score', hue='Category', dodge=True, alpha=0.7, edgecolor='black')
plt.xlabel("Country", fontsize=14)
plt.ylabel("Happiness Score", fontsize=14)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
plt.title("Comparison of Top Happiness Scores: EU vs MEA", fontsize=16)
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='best')

plt.tight_layout()
plt.savefig("top_comparison.png", dpi=300)
plt.show()
```

```python
# new graph for bottom
# Create figure for bottom comparison
fig, ax = plt.subplots(figsize=(12, 8))
sns.barplot(data=bottom_melted, x='Country', y='Happiness Score', hue='Category', dodge=True, alpha=0.7, edgecolor='black')
plt.xlabel("Country", fontsize=14)
plt.ylabel("Happiness Score", fontsize=14)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
plt.title("Comparison of Bottom Happiness Scores: EU vs MEA", fontsize=16)
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='best')
plt.tight_layout()
#plt.savefig("bottom_comparison.png", dpi=300)
plt.show()
```
